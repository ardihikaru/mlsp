```python
# Source: https://stackoverflow.com/questions/42150965/how-to-plot-gamma-distribution-with-alpha-and-beta-parameters-in-python
import numpy as np
import scipy.stats as stats
from matplotlib import pyplot as plt

'''
According to the documentation, you want to use the scale parameter (theta), 
but since you are defining beta, which is the inverse of theta, 
then you pass scale with the value of 1/beta, 
which in your example would be 1/3 or 0.33333.
'''

x = np.linspace (0, 100, 200) 
y1 = stats.gamma.pdf(x, a=29, scale=0.33333)
# y1 = stats.gamma.pdf(x, a=29, loc=3) #a is alpha, loc is beta???
# x = np.linspace (0, 20, 1) 
# print(x)
# alpha = 1
# beta = 100
# target_beta = 1 / beta
# y1 = stats.gamma.pdf(x, a=alpha, scale=target_beta)
# y1 = stats.gamma.pdf(x, a=29, loc=3) #a is alpha, loc is beta???
plt.plot(x, y1, "y-", label=(r'$\alpha=29, \beta=3$')) 
plt.ylim([0,0.08])
plt.xlim([0,150])
plt.show()
```

```python
import seaborn as sns
from scipy.stats import gamma
data_gamma = gamma.rvs(a=1, size=10)
x = np.arange(1, 21, 1)
print(data_gamma)
ax = sns.distplot(data_gamma,
                  kde=True,
                  bins=100,
                  color='black',
                  # color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Gamma Distribution', ylabel='Frequency')
```
